```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('PlayerPersonalData.csv', low_memory=False)
```

Гистограмма потенциала указывает на доминирующий потенциал и особенности распределения

```python
df.Potential.plot.hist(facecolor='purple', grid=True)
plt.title('Player potential histogram')
plt.xlabel('Рotential')
plt.ylabel('Frequency')
plt.savefig("histogram.png")
```

Из гистограммы видно, что наибольшее количество игроков с потенцилом раным 70-73
